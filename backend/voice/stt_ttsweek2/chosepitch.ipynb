## Edge-TTS

เป็นการลองปรับค่า rate,pitch ของเสียงไปเรื่อยๆจนกว่าจะเจอค่าที่มันดี

```python
import time 
import pandas as pd
import os
import asyncio
import edge_tts
from mutagen.mp3 import MP3
```

```python
df = pd.read_csv("processed_dialogue_v2.csv")
print("This is test data")
df.head(3)
```

```python
text = df["TTS Script"]
print(text)
```

```python
# adjustment
async def generate_test_voices(df, rate="+0%", pitch="+0Hz"):
    voice = "th-TH-PremwadeeNeural"
    
    # วนลูปเฉพาะ 5 แถวแรกก่อนก็ได้เพื่อความเร็วในการเทส
    for index, row in df.iterrows():
        text = row['TTS Script']
        # ตั้งชื่อไฟล์ให้รู้ว่ามาจากแถวไหน และใช้ค่า rate/pitch เท่าไหร่
        filename = f"test_row{index}_r{rate}_p{pitch}.mp3"
        
        communicate = edge_tts.Communicate(text, voice, rate=rate, pitch=pitch)
        await communicate.save(filename)
        print(f"Generated: {filename}")

# วิธีรันใน Jupyter (เนื่องจากเป็น async)
# ลองปรับค่าตรงนี้ไปเรื่อยๆ
await generate_test_voices(df, rate="+3%", pitch="+10Hz")
```

## gTTS

ปรับได้ไม่เยอะเอาเท่าที่ดี

```python
from gtts import gTTS
import pandas as pd

def generate_gtts_test(df, slow=False):
    for index, row in df.iterrows():
        text = row['TTS Script']
        filename = f"gtts_row{index}_slow{slow}.mp3"
        
        # gTTS ไม่ต้องใช้ await นะครับ
        tts = gTTS(text=text, lang='th', slow=slow)
        tts.save(filename)
        print(f"Generated: {filename}")

# วิธีรัน (ไม่ต้องมี await ข้างหน้า)
generate_gtts_test(df, slow=False)
```
